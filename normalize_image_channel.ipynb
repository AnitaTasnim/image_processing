#calculate mean per channel


To calculate the mean for each channel (R, G, and B), you need to collapse the spatial dimensions (height and width) while keeping the channel dimension intact. You can do this by reshaping the tensor and then calculating the mean. In code:

```python
# Assuming you have a tensor 'imgs' with shape (3, 6, 6, 5)
channel_means = dataset_new.view(3, -1).mean(dim=1)

# 'channel_means' will be a tensor of shape (3,) where each element represents the mean value for one channel (R, G, B)
```

```python
channel_means
```

#calculate standard deviation per channel

```python
channel_stds = dataset_new.view(3, -1).std(dim=1)
channel_stds
```
